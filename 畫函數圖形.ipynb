```python
%matplotlib inline   
#魔術指令%讓圖形直接顯示在網頁上
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x1= np.linspace(10e-7,10,1000)
x2= np.linspace(-10,10e-7,1000)
```

```python
plt.plot(x1,np.log2(x1),"r",
         x1,2**x1,"b",
         x2,2**x2,"b")
plt.axis([-10,10,-10,10])
plt.show()
```
